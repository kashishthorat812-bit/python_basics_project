```python
def analyze_string(s):
    if not s:
        print("Empty string!")
        return

    print("Length:", len(s))
    print("Reverse:", s[::-1])

    vowels = "aeiou"
    count = 0

    for ch in s.lower():
        if ch in vowels:
            count += 1

    print("Vowels:", count)

    for i in range(len(s)):
        print(f"Index: {i}, Negative Index: {i-len(s)}, Character: {s[i]}")

text = input("Enter a string = ")
analyze_string(text)
```
