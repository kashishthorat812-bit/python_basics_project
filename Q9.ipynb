```python
import math

try:
    sentence = input("Enter a sentence: ")

    words = set(sentence.lower().split())

    print("Unique Words:", sorted(words))

    total = len(words)
    print("Total Unique Words:", total)

    print("Square of Total Unique Words:", math.pow(total, 2))

except Exception as e:
    print("Error:", e)
```
