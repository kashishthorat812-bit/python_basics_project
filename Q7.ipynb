```python
square = lambda x: x * x

try:
    numbers = range(1, 21)

    squares = list(map(square, numbers))

    print("Squares:", squares)

    even_squares = [i for i in squares if i % 2 == 0]

    print("\nEven Squares:", even_squares)

except Exception as e:
    print("Error:", e)
```
